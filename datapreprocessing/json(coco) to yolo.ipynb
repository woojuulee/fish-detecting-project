# 공모전에서 제공한 데이터셋 형식(COCO)을 YOLO 데이터셋 형식으로 변환

```python
import json
import cv2
import os
# import matplotlib.pyplot as plt
import shutil
```

```python
input_path = "dataset"
output_path = "dataset/yolo_dataset/images"
```

```python
# Reading JSON Annotation file

f = open('original_dataset_1031/labels/train.json')
data = json.load(f)
f.close()
```

```python
def get_img(filename):
  for img in data['images']:
    if img['file_name'] == filename:
      return img
    
def get_img_ann(image_id):
    img_ann = []
    isFound = False
    for ann in data['annotations']:
        if ann['image_id'] == image_id:
            img_ann.append(ann)
            isFound = True
    if isFound:
        return img_ann
    else:
        return None
```

```python
#원본 데이터 다시 라벨링하면서 yolo data 형식으로 바꾸는 코드
from tqdm import tqdm
import os

def json2yolo(input_path,output_path):
    cnt = 0
    file_names = []

    for filename in tqdm(os.listdir(input_path)):
        
        source = os.path.join(input_path, filename)
        
        if get_img_ann(get_img(filename)['id']) is not None:
            try:
                # anno_over1 데이터의 경우 image를 따로 분리할 필요는 없기 때문에 아래 코드 주석처리.
                # destination = f"{output_path}/images/{filename}"
                # shutil.copy(source, destination)
                file_names.append(filename)
                
                cnt+=1
            except shutil.SameFileError:
                print("source and destination represents the same file")
    ######################################################################

    for filename in tqdm(file_names):
        # Extracting image 
        img = get_img(filename)
        img_id = img['id']
        img_w = img['width']
        img_h = img['height']

        # Get Annotations for this image
        img_ann = get_img_ann(img_id)

        if img_ann:
            # Opening file for current image
            name = filename.split(".")[0]

            # 기존 코드 주석처리
            # file_object = open(f"{output_path}/labels/{name}.txt", "a")

            # anno_over1 데이터의 경우, 이미지 파일과 같은 폴더에 라벨 데이터 위치시키기 위해 코드 수정.
            file_object = open(f"{output_path}/{name}.txt", "a")
            for ann in img_ann:
                
                current_category = ann['category_id'] 
                current_bbox = ann['bbox']
                x = current_bbox[0]
                y = current_bbox[1]
                w = current_bbox[2]
                h = current_bbox[3]
                
                # Finding midpoints
                x_centre = (x + (x+w))/2
                y_centre = (y + (y+h))/2
                # print(x_centre, y_centre, w, h)

                # Normalization
                x_centre = x_centre / img_w
                y_centre = y_centre / img_h
                w = w / img_w
                h = h / img_h
                print(x_centre, y_centre, w, h)

                # Limiting upto fix number of decimal places
                x_centre = format(x_centre, '.9f')
                y_centre = format(y_centre, '.9f')
                w = format(w, '.9f')
                h = format(h, '.9f')
                    
                # Writing current object 
                if len(img_ann) == 1 :
                    file_object.write(f"{current_category} {x_centre} {y_centre} {w} {h}")
                else :
                    file_object.write(f"{current_category} {x_centre} {y_centre} {w} {h}\n")

                


            file_object.close()
            # count += 1  # This should be outside the if img_ann block.
    return file_names
```

```python
input_path = "anno_over1"
output_path = "anno_over1"
json2yolo(input_path,output_path)
```
