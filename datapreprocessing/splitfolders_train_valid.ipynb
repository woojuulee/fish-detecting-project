# 폴더 안에 같이 있는 image와 label을 나누기

```python
import shutil
import os
# input folder를 original datafolder와 augment data folder로 변경해주면 된다.
input_folder = r"C:\Users\User\fish\data_yolov8_ver3_original" 
output_folder = r"C:\Users\User\fish\data_yolov8_ver3_test"

if not os.path.exists(output_folder):
    os.makedirs(output_folder)

for folder in os.listdir(input_folder):
    folder_path = os.path.join(input_folder,folder)
    for class_folder in os.listdir(folder_path): #class 반복
        class_folder_path = os.path.join(folder_path,class_folder)
        for file_name in os.listdir(class_folder_path):
            if file_name.endswith(".png") or file_name.endswith(".jpg"):
                image_source_path = os.path.join(class_folder_path, file_name)

                output_image_path = os.path.join(output_folder,folder,"images")
                if not os.path.exists(output_image_path):
                    os.makedirs(output_image_path)

                image_destination_path = os.path.join(output_image_path,file_name)
                shutil.copy(image_source_path, image_destination_path)
                print(f"Copied {file_name} to {output_image_path} 이미지파일 잘 옮겨갔어요~")
                
            elif file_name.endswith(".txt"):
                label_source_path = os.path.join(class_folder_path, file_name)

                output_label_path = os.path.join(output_folder,folder,"labels")
                if not os.path.exists(output_label_path):
                    os.makedirs(output_label_path)

                label_destination_path = os.path.join(output_label_path,file_name)
                shutil.copy(label_source_path, label_destination_path)
                print(f"Copied {file_name} to {output_label_path} 라벨파일 잘 옮겨갔어요~")
```
