```python
import json
import cv2
import os
# import matplotlib.pyplot as plt
import shutil

f = open('original_dataset_1031/labels/train.json')
data = json.load(f)
f.close()
```

```python
def get_img(filename):
  for img in data['images']:
    if img['file_name'] == filename:
      return img
    
def get_img_ann(image_id):
    img_ann = []
    isFound = False
    for ann in data['annotations']:
        if ann['image_id'] == image_id:
            img_ann.append(ann)
            isFound = True
    if isFound:
        return img_ann
    else:
        return None
```

```python
# 우선, 어노테이션이 있는 데이터만 select--> 원래 이름 그대로 들고옴
from tqdm import tqdm
import os

def load_images_from_folder(folder):
    output_path = "./dataset_ver6"

    for filename in tqdm(os.listdir(folder)):
        
        source = os.path.join(folder, filename)
        destination = os.path.join(output_path, filename)
        if get_img_ann(get_img(filename)['id']) is not None:
            try:
                shutil.copy(source, destination)
            except shutil.SameFileError:
                print("source and destination represents the same file")
```

```python
folder = r'C:\Users\User\fish\original_dataset_1031\images\train'
load_images_from_folder(folder)
```
